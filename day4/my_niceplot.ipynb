```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
data = np.random.normal(loc=0., scale=5., size=(50000,))
from scipy.stats import norm
a,b = norm.fit(data)
```

```python
fig = plt.figure()
ax  = fig.add_subplot(111)
ax.set_title('My nice plot')
ax.hist(data, range=(-15,15), bins=100, density=True, label='random data',color='c',lw=0.2)
ax.plot(np.arange(-15,15,0.1),norm.pdf(np.arange(-15,15,0.1),a,b), label='fit', color='g',linewidth=4)
ax.set_xlabel(r'Numpy data')
ax.set_ylabel(r'Distribution, m')
ax.legend(loc=1)
plt.show()
```
