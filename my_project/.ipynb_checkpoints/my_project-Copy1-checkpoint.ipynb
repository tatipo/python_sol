```python
# originalMatlab.m to python 3

import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
```

```python
# view raw data file
input_f = 'Islandsberg%7D2015-12-24T00h00A.1rw'
with open(input_f,'r') as reader:
    print(reader.read())
```

```python
# fetch second column as Wave elevation in heave
column1 = [] 
with open(r"Islandsberg%7D2015-12-24T00h00A.1rw", "r+") as f: 
    data = f.readlines() 
    for row in data: 
        column1.append(row) 
print(column1)
```

```python
# convert list to numpy array integers
arr=np.array(column1,dtype=int)

print(arr)
```

```python
# convert to numpy and convert from cm to m
arr=arr/100
print(arr)
```

```python
# get the wave
wave=arr-np.mean(arr)
lenW=wave.size

print(wave)
print(lenW)
plt.plot(wave)
plt.xlim([800,1000])
plt.ylabel('Wave elevation [m]')
plt.show()
```

```python
# define sampling frequency and time vector
fs=2.56
dt=1/fs
t=np.arange(dt,lenW*dt,dt)
print(t)
```

```python
# compute spectral density with fast fourier transform
z=np.fft.fft(wave)/lenW
df=1/lenW/dt
Nyf=round(z.size/2)
f=np.arange(df,df*Nyf,df)
Sp=2*(z[1:Nyf])*np.conj(z[1:Nyf])/df
S=np.real(Sp)
print(S)

plt.plot(f,S)
plt.xlim([0,0.5])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Spectral density [m^2 s]')
plt.show()
```

```python
# compute spectral moments and wave paramerters using trapesoidal integration
m0=np.trapz(f,S)
mm1=np.trapz(f,S/f)
m1=np.trapz(f,S*f)
m2=np.trapz(f,S*f**2)
m4=np.trapz(f,S*f**4)

Hm0=4*np.sqrt(np.absolute(m0))
print(Hm0)
print('Significant wave height is ',Hm0,' m')
Te=mm1/m0
print('Energy period is ',Te, ' s')
Tavg=m0/m1
print('Average period is ',Tavg,' s')
epsilon=np.sqrt(1-m2**2/m0/m4)
print('Bandwidth is ',epsilon)
Tz=np.sqrt(m0/m2)
print('Average up crossing period is ',Tz,'s')
Tp=np.sqrt(m2/m4)
print('Peak period is ',Tp, ' s')
ind=np.argmax(S)
fm=f[ind]
Tm=1/fm
print('Peak spectrum period is ',Tm, ' s')
```

```python
# calculate energy flux
rho=1025
g=9.81
J=rho*g**2/64/np.pi*Te*Hm0**2
print('Energy flux is ',J,'W/m')
```

```python
# save the results
resultWaveData = np.zeros(8)
resultWaveData[0] = Hm0
resultWaveData[1] = Te
resultWaveData[2] = Tavg
resultWaveData[3] = Tz
resultWaveData[4] = Tp
resultWaveData[5] = Tm
resultWaveData[6] = epsilon
resultWaveData[7] = J
print(resultWaveData)
```

```python
# save as txt
newF = open('NewWaveData.txt','w')
np.savetxt(newF,resultWaveData,delimiter=',')

newF.close()
```
